```python
import pickle
import numpy as np

alphas = 10**(-1 * np.arange(0, 4, 1.0))
l1_ratios = 10**(-1 * np.arange(0, 3, 1.0/3))
hyperlist = zip(list(np.repeat(alphas, len(l1_ratios))), list(np.tile(l1_ratios, len(alphas))))

filename = "hyperlist_enet.p"


print "Hyper list is "
print hyperlist

pickle.dump(hyperlist, open(filename, 'wB'))

print "Dumped to ", filename
print "Loading: "
pickle.load(open(filename, 'rB'))
```
