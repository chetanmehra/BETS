```python
import pickle
import numpy as np

alphas = 10.0**(np.arange(-4, 1))
l1_ratios = np.arange(0.1, 1.0, 0.2)
hyperlist = zip(list(np.repeat(alphas, len(l1_ratios))), list(np.tile(l1_ratios, len(alphas))))

filename = "hyperlist_enet.p"


print "Hyper list is "
print hyperlist

pickle.dump(hyperlist, open(filename, 'wB'))

print "Dumped to ", filename
print "Loading: "
pickle.load(open(filename, 'rB'))
```
