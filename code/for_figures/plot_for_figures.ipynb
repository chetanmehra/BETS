# For Bootstrap Edge Frequencies

```python
# Scratch work

import numpy as np
import os

np.random.seed(121)

u = np.random.normal(loc=12, scale=1.5, size=100)
s = np.random.normal(loc=10, scale=2.0, size=100)
bins = np.linspace(6, 16, 15)

import matplotlib.pyplot as plt
%matplotlib inline



filename = os.path.join("plots", "bootstrap-edge-fdr-control.pdf")

fig = plt.figure()
plt.hist(u, alpha=0.5, color = 'red', label="Original", bins=bins )
plt.hist(s, alpha=0.5, color = 'blue', label="Permuted", bins=bins)
plt.xlim(4,20)
plt.ylim(0, 24)
plt.xlabel("Bootstrap Edge Frequency", fontsize=20)
plt.axvline(x = 14, label="Threshold", color='black', linestyle='dashed', linewidth=3.0)
plt.title("FDR Control", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.legend()

fig.savefig(filename)
print "Saved to ", filename
plt.show()
```
